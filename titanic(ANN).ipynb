```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

titanic_train=pd.read_csv('C:\\Users\\hemanth kumar\\Desktop\\kaggle project\\titanic\\train.csv')
titanic_test=pd.read_csv('C:\\Users\\hemanth kumar\\Desktop\\kaggle project\\titanic\\test.csv')

y_train=titanic_train.iloc[:,1].values

titanic_train=titanic_train.drop(['PassengerId','Survived','Name','Ticket','Cabin'],axis=1)
titanic_test=titanic_test.drop(['PassengerId','Name','Ticket','Cabin'],axis=1)


x_train=titanic_train.iloc[:,:].values
x_test=titanic_test.iloc[:,:].values

x=np.vstack((x_train,x_test))

X=pd.DataFrame(x,columns=['Pclass', 'Sex', 'Age', 'SibSp','Parch','Fare', 'Embarked'])

X['Embarked'].fillna(value='S',inplace=True)
X['Age'].fillna(value=X['Age'].mean(),inplace=True)
X['Fare'].fillna(value=X['Fare'].mean(),inplace=True)
```

```python
X.head(10)
```

```python
X.isnull().values.any()
```

```python
X=X.iloc[:,:].values
X[1,:]
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
le=LabelEncoder()
X[:,1]=le.fit_transform(X[:,1])

from sklearn.compose import ColumnTransformer
ct=ColumnTransformer(transformers=[('ohe',OneHotEncoder(),[-1])],remainder='passthrough')

X=ct.fit_transform(X)
```

```python
X=X[:,1:]
```

```python
X[1,:]
```

```python
X=pd.DataFrame(X,columns=['e1','e2','Pclass','Sex','Age','SibSp','Parch','Fare'])
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
ss_x=StandardScaler()
X=ss_x.fit_transform(X)
# now x_train and x_test are arrays
```

```python
x_train=X[0:891,:]
x_test=X[891:,:]
```

```python
x_train[1,:]
```

## neural network

```python
import keras 
from keras import Sequential
from keras.layers import Dense

classifier = Sequential([Dense(input_dim=8,output_dim=5,init='uniform',activation='relu'),
                        Dense(output_dim=5,init='uniform',activation='relu'),
                        Dense(output_dim=5,init='uniform',activation='relu'),
                        Dense(output_dim=1,init='uniform',activation='sigmoid')])
```

```python
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
classifier.fit(x_train,y_train,batch_size=5,epochs=30)
```

```python
y_pred=classifier.predict(x_test)
y_pred=(y_pred>0.5)

y_test=pd.read_csv('C:\\Users\\hemanth kumar\\Desktop\\gender_submission (4).csv')
y_test.head()

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test.iloc[:,1],y_pred)
cm
```
