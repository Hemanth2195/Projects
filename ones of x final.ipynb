## applying BFGS

```python
from scipy.optimize import minimize
```

```python
def mul(a,b):
    return np.dot(a,b)
```

```python
def j_theta(theta1):
    theta1=pd.DataFrame(theta1,index=x.columns)
    h_theta1=np.dot(x,theta1)
    h_theta1=pd.DataFrame(h_theta1)
    h_theta1=1/(1+np.exp(-h_theta1))
    m=len(x)
    b=(-1/m)*(mul(y.T,np.log(h_theta1))+mul((1-y).T,np.log(1-h_theta1)))
    return b
```

```python
result=minimize(j_theta,theta,method='BFGS')
```
